Problem 5

Get all items of ser1 and ser2 not common to both.

Input

```
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
```

````python
ser1, ser2 = pd.Series([1, 2, 3, 4, 5]), pd.Series([4, 5, 6, 7, 8]) 
# This line creates two Pandas Series objects ser1 and ser2, each with a list of integer values.
ser11 = pd.Series(np.union1d(ser1, ser2))
"""This line uses the np.union1d() function to create a new Series object ser11 that contains the union of the elements in ser1 and ser2. 
The resulting Series contains all the unique elements from both ser1 and ser2."""
#print(sr11)
sr22 = pd.Series(np.intersect1d(ser1, ser2))
"""This line uses the np.intersect1d() function to create a new Series object sr22 that contains the intersection of the elements in ser1 and ser2.
The resulting Series contains only the elements that are present in both ser1 and ser2."""
# print(sr22)
result = sr11[~sr11.isin(sr22)] 
#This line prints the resulting Series result, which contains only the elements from ser1 and ser2 that are unique to either ser1 or ser2.
print(result)
````

Get the day of month, week number, day of year and day of week from ser.

Input
```

```
Desired output
```
Date:  [1, 2, 3, 4, 5, 6]
Week number:  [53, 5, 9, 14, 19, 23]
Day num of year:  [1, 33, 63, 94, 125, 157]
Day of week:  ['Friday', 'Wednesday', 'Saturday', 'Thursday', 'Monday', 'Saturday']
```

````python
# define a pandas Series containing a list of date strings in different formats
ser = pd.Series(['01 Jan 2010', '02-02-2011', '20120303', '2013/04/04', '2014-05-05', '2015-06-06T12:20'])
# create an empty dictionary to store the extracted date components
dic = {}
# create a pandas DataFrame from the Series, with a single column named "time"
time = pd.DataFrame(ser, columns=["time"])
# convert the "time" column of the DataFrame to pandas datetime format
time['time'] = pd.to_datetime(time['time'])
# extract the day, week number, day number of year, and day of week for each date
dic["Date"] = time['time'].dt.day.tolist()
dic["Week number"] = time['time'].dt.week.tolist()
dic["Day num of year"] = time['time'].dt.dayofyear.tolist()
dic["Day of week"] = time['time'].dt.day_name().tolist()
# print the dictionary
for key, value in dic.items():
    print(key + ": ", value)
````

Problem 19

In df, use apply method to replace the missing values in Min.Price with the column’s mean and those in Max.Price with the column’s median.

Input

```
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')

```

````python
# load the dataset from the specified URL
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')

# define a function to fill in missing values
def a(col):
  if col.name == 'Min.Price':  # if the column is 'Min.Price'
    return col.fillna(col.mean())  # fill missing values with the mean of the column
  elif col.name == 'Max.Price':  # if the column is 'Max.Price'
    return col.fillna(col.median())  # fill missing values with the median of the column
  else:  # for any other column
    return col  # return the column as is

# apply the function to the 'Min.Price' and 'Max.Price' columns using the apply() method
# and axis=0 to apply the function to each column separately
df[['Min.Price', 'Max.Price']] = df[['Min.Price', 'Max.Price']].apply(a, axis=0)

# output information about the 'Min.Price' and 'Max.Price' columns
df[['Min.Price', 'Max.Price']].info()
````

````python
# 2 - Create a generic function to interchange two columns, without hardcoding column names.
# Define a function to interchange two columns in a DataFrame, given their names
def interchange_columns(df, col1, col2):
    # Get the current column order
    current_order = df.columns.tolist()
    # Get the indices of the columns to interchange
    idx1 = current_order.index(col1)
    idx2 = current_order.index(col2)
    # Swap the column positions in the list
    current_order[idx1], current_order[idx2] = current_order[idx2], current_order[idx1]
    # Return a new DataFrame with the columns in the new order
    return df[current_order]
# Call the interchange_columns function to swap the 'd' and 'a' columns in df
df02 = interchange_columns(df, 'd', 'a')
# Print the resulting DataFrame
df02
````
