### combining and reshaping data

```python
customers = {"order_id":[11,12],"c_id":[1,2],"name":["a","b"],"add":["kar","hyd"]}
customers = pd.DataFrame(customers)
```

```python
import datetime
orders = {"c_id":[1,2,3],
          "data_order":[datetime.date(2023,8,9),datetime.date(2023,9,5),datetime.date(2023,9,10)],}
orders = pd.DataFrame(orders)
orders
```

```python
mar_data=pd.merge(customers,orders,how="inner")
# inner >> intersection 
mar_data
```

```python
mar_data=pd.merge(customers,orders,how="outer")
# outer >> union 
mar_data
```

```python
mar_data=pd.merge(customers,orders,how="outer")
# outer >> union 
mar_data.fillna("")
```

```python
mar_data
```

```python
mar_data=pd.merge(customers,orders,how="left")
# left has most important right now 
mar_data
```

```python
mar_data=pd.merge(customers,orders,how="right")
mar_data
```

```python
df1 = pd.DataFrame({'left': ['foo', 'bar']})
df2 = pd.DataFrame({'right': [7, 8]})
df = pd.merge(df1,df2, how='cross')
df
```

```python
df1 = pd.DataFrame({'key':[1,1],'left': ['foo', 'bar']})
df2 = pd.DataFrame({'key':[1,1],'right': [7, 8]})
df = pd.merge(df1,df2, how='cross')[['left','right']]
df
```
