```python
import pandas as pd
import numpy as np
#hide earrings 
import warnings
warnings.filterwarnings('ignore')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD
#from tensorflow.keras.utils import to_categorical
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dataset/iris.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
print(f"{df.info()}\n{df.shape}")
```

```python
def get_seprate_features(df_id,col_id):
    #col_id is special col or target col b/c this the whole model is going to be created 
    global X, y
    X = df_id.drop(col_id,axis = 1) 
    y = df_id[col_id]
get_seprate_features(df,['species'])
```

```python
X
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
X = min_max_scaler.fit_transform(X)
X
```

```python
y
```

```python
y.shape
```

```python
encoder =  LabelEncoder()
y = encoder.fit_transform(y)
y.shape
```

```python
y = pd.get_dummies(y)#.values
```

```python
y
```

```python
def get_train_test_split(array1, array2, test_size, random_state):
    global X_train, X_test, y_train, y_test
    X_train, X_test, y_train, y_test = train_test_split(array1, array2, test_size=test_size,random_state=random_state)
get_train_test_split(X,y,0.3,1)
```

```python
print(f'X_train: {X_train.shape} y_train: {y_train.shape}  \nX_test: {X_test.shape} y_test: {y_test.shape}')
```

```python
model = Sequential()
model.add(Dense(32, activation='relu', input_shape=(4,)))
model.add(Dense(24, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=50, epochs=100)
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test) # evaluate -> estimate the significance
```

```python
print(f'Test accuracy:{test_acc*100}\nTest loss: {test_loss*100}')
```
